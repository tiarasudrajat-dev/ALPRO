```python
import math

def fase_turun(m, sudut):
    g = 9.8
    # math.sin membutuhkan input dalam radian, jadi kita konversi dari derajat
    return m * g * math.sin(math.radians(sudut))

def fase_naik(m, sudut):
    # Logika sesuai kode Pascal: FaseTurun + (m * g * sin(theta))
    # Ini secara efektif menghitung gaya gravitasi paralel yang harus dilawan
    g = 9.8
    return fase_turun(m, sudut) + (m * g * math.sin(math.radians(sudut)))

def hitung_daya(f, v):
    return f * v

def tampilkan_proses(m_in, v_in, t_in):
    f_fr = fase_turun(m_in, t_in)
    f_p = fase_naik(m_in, t_in)
    p_hasil = hitung_daya(f_p, v_in)

    print("\n>>>> LANGKAH PERHITUNGAN <<<<")
    print(f"a. f_fr (Fase Turun) : {f_fr:.2f} N")
    print(f"b. Fp   (Fase Naik)  : {f_p:.2f} N")
    print(f"c. P    (Rumus Daya) : {f_p:.2f} * {v_in:.2f}")
    print("-" * 29)
    print(f"HASIL DAYA AKHIR     : {p_hasil:.2f} Watt")
    print("-" * 29)

def main():
    print("=== ANALISIS DAYA SESUAI CATATAN ===")
    try:
        m = float(input("Massa (kg)     : "))
        v = float(input("Kecepatan (m/s): "))
        theta = float(input("Sudut (deg)    : "))

        tampilkan_proses(m, v, theta)
    except ValueError:
        print("Input harus berupa angka!")

    input("\nTekan Enter untuk keluar...")

if __name__ == "__main__":
    main()
```
